```python
classe=[
    ["Alice", 20, 15.5],
    ["Bob", 19, 12.0],
    ["Charlie", 22, 16.0]
]
```

```python
def afficher_classe(classe):
    if not classe:
        print("La classe est vide.")
        return
    print("\nListe des etudiants: ")
    for index,(nom,age,note) in enumerate(classe,start=1):
        print(f"{index}.{nom} - {age} ans - note {note}")
    
```

```python
def ajouter_etudiant(classe):
    nom=input("Nom: ").strip()
    if not nom:
        print("Nom vide, ajout annulé.")
        return
    try:
        age=int(input("Age : ").strip())
        note=float(input("Note :").strip())
    except ValueError:
        print("Saisie invalide")
        return
    classe.append([nom,age,note])
    print(f"{nom} ajouté")
```

```python
def supprimer_etudiant(classe):
    if not classe:
        print("Rien à supprimer")
        return

    try:
        index = int(input("num de l'etudiant à supprimer : ").strip()) - 1

        if index < 0 or index >= len(classe):
            raise IndexError

        etudiant = classe[index]

        confirmation = input(f"Confirmer la suppression de {etudiant[0]} ? (o/n) : ").strip().lower()
        if confirmation != "o":
            print("Suppression annulée.")
            return  

        classe.pop(index)
        print(f"{etudiant[0]} supprimé.")

    except ValueError:
        print("Veuillez entrer un nombre valide.")
    except IndexError:
        print("Index invalide : aucun étudiant avec ce numéro.")
```

```python
def mettre_a_jour_etudiant(classe):
    if not classe:
        print("Classe vide, rien à modifier.")
        return

    try:
        index = int(input("Numéro de l’étudiant à modifier : ").strip()) - 1
        etudiant = classe[index]
    except ValueError:
        print("Veuillez entrer un nombre valide.")
        return
    except IndexError:
        print("Index invalide.")
        return

    print(f"\nModification de {etudiant[0]} (laisser vide pour ne pas changer).")

    nouveau_nom = input("Nouveau nom : ").strip()
    if nouveau_nom:
        etudiant[0] = nouveau_nom

    entree_age = input("Nouvel âge : ").strip()
    if entree_age:
        try:
            etudiant[1] = int(entree_age)
        except ValueError:
            print("Âge ignoré (saisie invalide).")

    entree_note = input("Nouvelle note : ").strip()
    if entree_note:
        try:
            etudiant[2] = float(entree_note)
        except ValueError:
            print("Note ignorée (saisie invalide).")

    print("Étudiant mis à jour :", etudiant)
```

```python
def afficher_statistiques(classe):
    if not classe:
        print("Pas de données (classe vide).")
        return

    notes = [etudiant[2] for etudiant in classe]
    moyenne = sum(notes) / len(notes)
    meilleure = max(classe, key=lambda e: e[2])
    pire = min(classe, key=lambda e: e[2])

    print("\n--- Statistiques ---")
    print(f"Moyenne des notes : {moyenne:.2f}")
    print(f"Meilleure note : {meilleure[2]} ({meilleure[0]})")
    print(f"Moins bonne note : {pire[2]} ({pire[0]})")
```

```python
def rechercher_par_nom(classe):
    if not classe:
        print("Classe vide.")
        return

    terme = input("Nom à rechercher (partiel possible) : ").strip().lower()
    if not terme:
        print("Terme vide, recherche annulée.")
        return

    resultats = []
    for index, (nom, age, note) in enumerate(classe, start=1):
        if terme in nom.lower():
            resultats.append((index, nom, age, note))

    if not resultats:
        print("Aucun étudiant trouvé pour ce nom.")
    else:
        print("\nRésultats de la recherche :")
        for idx, nom, age, note in resultats:
            print(f"{idx}. {nom} — {age} ans — note {note}")
```

```python
while True:
    print("\n===== Gestion de la classe =====")
    print("1. Afficher tous les étudiants")
    print("2. Ajouter un étudiant")
    print("3. Supprimer un étudiant")
    print("4. Modifier un étudiant")
    print("5. Statistiques")
    print("6. Rechercher par nom")
    print("q. Quitter")

    choix = input("Choix : ").strip().lower()

    if choix == "1":
        afficher_classe(classe)
    elif choix == "2":
        ajouter_etudiant(classe)
    elif choix == "3":
        supprimer_etudiant(classe)
    elif choix == "4":
        mettre_a_jour_etudiant(classe)
    elif choix == "5":
        afficher_statistiques(classe)
    elif choix == "6":
        rechercher_par_nom(classe)
    elif choix == "q":
        print("Au revoir.")
        break
    else:
        print("Option inconnue, merci de réessayer.")
```
